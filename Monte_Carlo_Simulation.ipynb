#Monte Carlo Simulation of a simple project


```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_simulations = 1000

#Cost data
#Assumption: x1, x2 and x3 are uniformly distributed, x4 is normally distributed
#
x1_min = 5000
x1_max = 30000
x2_min = 50000
x2_max = 100000
x3_min = 15000
x3_max = 45000
x4_mean = 50000
x4_std_dev = 25000
Fixed_cost = 50000
```

```python
x1 = []
x2 = []
x3 = []
x4 = []
total_cost = []

for i in range(num_simulations):
    x1_cost = (int)(x1_min+(x1_max-x1_min)*random.random())
    x2_cost = (int)(x2_min+(x2_max-x2_min)*random.random())
    x3_cost = (int)(x3_min+(x3_max-x3_min)*random.random())
    x4_cost = (int)(np.random.normal(x4_mean, x4_std_dev))
        
    #store the data
    x1.append(x1_cost)
    x2.append(x2_cost)
    x3.append(x3_cost)
    x4.append(x4_cost)

#calculate total cost
for i in range(len(x1)):
    total_cost.append(x1[i]+x2[i]+x3[i]+x4[i]+Fixed_cost)
```

```python
#calculate mean and standard deviation
total_cost_mean = (int)(np.average(total_cost))
total_cost_dev = (int)(np.std(total_cost))
```

```python
plt.hist(total_cost)
plt.show() 
```
